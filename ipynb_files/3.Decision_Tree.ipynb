# Importing essential libraries

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import pandas as pd
import numpy as np
from sklearn import metrics
```

# Load the dataset

```python
# Prepare the data data
iris = datasets.load_iris()
print(iris)
```

```python
X = iris.data
y = iris.target
```

# Split the data into training/testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

# Generating Model

```python
dt_first = DecisionTreeClassifier()
```

```python
dt_first.fit(X_train, y_train)
```

```python
y_pred = dt_first.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

# Evaluating the Model

```python
# numbers on the diagonal of the confusion matrix correspond to correct predictions

#species = np.array(y_test).argmax(axis=1)
#predictions = np.array(y_pred).argmax(axis=1)
confusion_matrix(y_test, y_pred)
```

```python
score1 =classification_report(y_test, y_pred)
print(score1)
```

```python
# List of values to try for max_depth:
max_depth_range = list(range(1, 6))
# List to store the accuracy for each value of max_depth:
accuracy = []
for depth in max_depth_range:
    
    df = DecisionTreeClassifier(max_depth = depth, 
                             random_state = 0)
    df.fit(X_train, y_train)
    y_pred = df.predict(X_test)
    score = df.score(X_test, y_test)
    accuracy.append(score)
```

```python
print(accuracy)
```

# Training the hyperparameter with Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_dist={
    "criterion":["gini","entropy"],
    "max_depth":[1,2,3,4,5]
}
```

```python
classifier=DecisionTreeClassifier()
```

```python
grid=GridSearchCV(classifier, param_grid=param_dist, cv=10, n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_) 
```

```python
grid_predictions = grid.predict(X_test) 
```

```python
  # print classification report 
print(classification_report(y_test, grid_predictions))
```

```python
print(confusion_matrix(y_test,grid_predictions)) 
```
